```python
# prompt: Menggunakan data DataFrame: buatkan code untuk  uji stasioneritas data Close, High. Low. Open

from statsmodels.tsa.stattools import adfuller

# Define the columns to test for stationarity
columns_to_test = ['Close', 'High', 'Low', 'Open']

# Loop through each column and perform the ADF test
for column in columns_to_test:
    # Extract the column data
    data_column = data[column]

    # Perform the Augmented Dickey-Fuller test
    result = adfuller(data_column)

    # Print the test results
    print(f"Results of Augmented Dickey-Fuller Test for {column}:")
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

    # Interpret the p-value
    p_value = result[1]
    if p_value <= 0.05:  # Common significance level
        print("The time series is likely stationary.\n")
    else:
        print("The time series is likely non-stationary.\n")
```

```python
# prompt: buatkan saya code untuk regresi var dengan lag optimum

import yfinance as yf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.api import VAR
from statsmodels.tools.eval_measures import rmse, aic
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Download historical data for BNB from June 1, 2024 to June 30, 2024
data = yf.download("BNB-USD", start="2024-06-01", end="2024-06-30")

# Select relevant columns for VAR analysis
data = data[['Close', 'High', 'Low', 'Open', 'Volume']]

# Function to determine optimal lag for VAR model
def optimal_lag_var(data, maxlags):
    results = []
    for i in range(1, maxlags + 1):
        model = VAR(data)
        result = model.fit(i)
        results.append([i, result.aic])  # Use AIC for lag order selection

    results_df = pd.DataFrame(results, columns=['Lag', 'AIC'])
    optimal_lag = results_df['Lag'][results_df['AIC'].idxmin()]
    return optimal_lag

# Find the optimal lag order
max_lags = 12 # Adjust maximum lag as needed
optimal_lag = optimal_lag_var(data, max_lags)
print(f"Optimal Lag Order: {optimal_lag}")

# Fit the VAR model with the optimal lag
model = VAR(data)
var_model_result = model.fit(optimal_lag)
print(var_model_result.summary())

# Forecast (example: forecast for the next 5 steps)
lag_order = optimal_lag
forecast_steps = 10
forecast = var_model_result.forecast(data.values[-lag_order:], forecast_steps)

# Create a proper DatetimeIndex for the forecast DataFrame
# This ensures the index is a continuation of the original data's index
forecast_index = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=forecast_steps, freq='D')

forecast_df = pd.DataFrame(forecast, columns=data.columns, index=forecast_index)
print("\nForecast:")
print(forecast_df)

# Evaluate model (example: RMSE)
# You would need to adjust this part based on how you want to evaluate the model
rmse_values = []
for col in data.columns:
    rmse_values.append(rmse(data[col].iloc[-forecast_steps:], forecast_df[col]))
rmse_df = pd.DataFrame({'Variable':data.columns,'RMSE': rmse_values})
print("\nRMSE:")
rmse_df
```
